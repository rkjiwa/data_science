# Iris Dataset KNN Implementation

```python
# imports

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# import accuracy_score
from sklearn.metrics import accuracy_score
```

```python
#import data
iris = pd.read_csv('iris.data', 
                   sep = ",",
                   header=None,
                   names = ['sepal_length','sepal_width','petal_length','petal_width','class'])
```

```python
#look at the head of the iris data
iris.head()
```

```python
# take a look at the data
iris.describe()
```

```python
# look at the info, check for null or objects
iris.info()
```

```python
# ensure that the class column only includes the 3 flowers names
iris['class'].unique()
```

```python
# Plot Sepal length vs sepal width
sns.lmplot(x='sepal_length', y='sepal_width', hue='class', data=iris);
plt.title('Sepal Width vs Length');
```

```python
# plot petal length vs petal width
sns.lmplot(x='petal_length', y='petal_width', hue='class', data=iris);
plt.title('Petal Width vs Length');
```

```python
#assign X and Y for sklean
X = iris.loc[:,:'petal_width']
Y = iris['class']
```

```python
# take a look at X set
X.head()
```

```python
# take a look at the Y set
Y.head()
```

```python
#test train split, to split the X and Y sets into training and test groups
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30)
```

```python
# take a look at the x training dataset
X_train.head()
```

```python
# taking a look at the Y training dataset
Y_train.head()
```

```python
# take a look at the x test dataset
X_test.head()
```

```python
# take a look at the y test dataset
Y_test.head()
```

```python
#initiate the model, specify that we are using KneighborsClassifier
model = KNeighborsClassifier()
```

```python
# fitting the model to the training data
model.fit(X_train, Y_train)
```

```python
#come up with predictions using the trained model 
X_Pred = model.predict(X_test)
```

```python
# take a look at the predictions
X_Pred
```

```python
# take a look at the actual results
Y_test.values
```

```python
# calculate the accuracy score of the model
accuracyScore = np.count_nonzero((X_Pred == Y_test.values)) / len(X_Pred)
accuracyScore
```

```python
# calculate the number of correct predictions
numberCorrectPredictions = np.count_nonzero((X_Pred == Y_test.values))
numberCorrectPredictions
```

```python
# number k
k = [1,3,5,7,10,20,30,40,50]

# average score for k
avgScore = []

# number correct prediction
numCorrectPrediction = []
```

```python
# this runs the KNN algorithm for different k values, generate 10 splits for each k value.
# It trains the model, then comes up with predictions and provides the accuracy score and the number of correct predictions as a list.
for i in k:
    for j in range(10):
        # specifies the model to be used and the k value
        model = KNeighborsClassifier(n_neighbors=i)
        # split the dataset into train and test
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
        # fits the model to the training set
        model.fit(X_train, Y_train)
        # predictions
        Y_pred = model.predict(X_test)
        # append mean to lists
        avgScore.append(accuracy_score(Y_test, Y_pred, normalize = True).mean())
        numCorrectPrediction.append(accuracy_score(Y_test, Y_pred, normalize = False).mean())
```

```python
# this is to take the avgScore list and the numCorrectPredictions list and split them into each of the 10 train test groups
def listSplit(l, n):
    for i in range(0, len(l), n):
        yield l[i:i+n]
```

```python
# split the average score into a list
avgScoreList = list(listSplit(avgScore, 10))
# create a df from the avgScoreList and tranpose it
df = pd.DataFrame(avgScoreList)
# add the k values as a column, make them the index and rename the columns
df['k'] = k
df.set_index('k', inplace=True)
df.rename({0:'Group 1', 1:'Group 2', 2:'Group 3', 3:'Group 4', 4:'Group 5',5:'Group 6', 6:'Group 7',
              7:'Group 8', 8:'Group 9',9:'Group 10'}, axis='columns', inplace=True)
# check out the data
df
```

```python
# plot accuracy score
plt.plot(df);
plt.rcParams["figure.figsize"] = (16,9)
plt.ylabel('Accuracy Score')
plt.xlabel('N')
plt.title('Accuracy Score Neighbors')
plt.legend(df);
```

The general trend in the above plot is that as the k values increase, the accuracy score of the algorithm tends to decrease. The explanation for this would that as you increase k, you are increasing the distance and examining further points. Introducing these further points increases the likelihood that you will be including data that belongs to a different class but has traits that are borderline between the classes. These points will influence the training of the model and alter the weighting, reducing the impact of the closer points. Overall, this decreases accuracy. This is a general trend and is not a hard and fast rule. It is possible to end up with instances where the higher dimension show better accuracy that the lower dimensions. This is due to the fact that we are randomly generating the data. However, in this instance, all the groups show a decrease in accuracy as k increases. You can also see this by taking a look at the table. Based off of this we can determine that this algorithm is better suited when you are looking at a smaller number of neighbours. If you are looking at using a higher dimension, it would be wiser to look at a more sophisticated model.
